```python
raw = """Player 1:
9
2
6
3
1

Player 2:
5
8
4
7
10"""
from collections import deque

def parse(raw):
    return [deque(int(c) for c in p.splitlines()[1:]) for p in raw.split('\n\n')]

p1, p2 = parse(raw)
p1, p2 = parse(open('data/22.txt').read())
```

```python
def round(p1, p2):
    c1 = p1.popleft()
    c2 = p2.popleft()
    if c1 > c2:
        p1.extend([c1,c2])
    else:
        p2.extend([c2,c1])
```

```python
while len(p1) > 0 and len(p2) > 0:
    round(p1, p2)
```

```python
p = p1 if len(p1) > 0 else p2
def score(p):
    return sum((i+1)*v for i,v in enumerate(list(p)[::-1]))

score(p)
```

```python
def state(p1, p2):
    return (tuple(p1), tuple(p2))
```

```python
def game(p1, p2, c=1):
    print('game', c)
    prev = set()
    r = 1
    while len(p1) > 0 and len(p2) > 0:
        if round(p1, p2, prev, g=c, r=r):
            break
        r += 1

    p1_wins = len(p1) > 0
    p = p1 if len(p1) > 0 else p2
    print('game', c, 'ended', p1_wins)
    return p1_wins, (score(p), p1, p2)
    
```

```python
def round(p1, p2, prev=set(), g=-1, r=1):
    #print('round', r, p1, p2)
    if state(p1, p2) in prev: # have seen this before
        print('break recursion')
        return True
    else:
        prev.add(state(p1,p2))
    c1 = p1.popleft()
    c2 = p2.popleft()
    if c1 <= len(p1) and c2 <= len(p2): ## start subgame
        new_p1 = deque(list(p1)[::][:c1])
        new_p2 = deque(list(p2)[::][:c2])
        p1_wins, _ = game(new_p1, new_p2, c=g+1)
    else:
        p1_wins = c1 > c2

    if p1_wins:
        p1.extend([c1,c2])
    else:
        p2.extend([c2,c1])
    
```
